```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

dataset = pd.read_csv("iris_dataset.csv")

figure1 = plt.figure(1, figsize=(20, 10))

chart1 = figure1.add_subplot(221)
chart2 = figure1.add_subplot(222)
chart3 = figure1.add_subplot(223)
chart4 = figure1.add_subplot(224)

chart1.scatter(dataset['flower_class'], dataset['sepal_len'])
chart1.set_xlabel("Class of flower")
chart1.set_ylabel("Sepal length (cm)")
chart1.set_title("Sepal length vs class")

chart2.scatter(dataset['flower_class'],dataset['sepal_wid'])
chart2.set_xlabel("Class of flower")
chart2.set_ylabel("Sepal width (cm)")
chart2.set_title("Sepal width vs class")

chart3.scatter(dataset['flower_class'],dataset['petal_len'])
chart3.set_xlabel("Class of flower")
chart3.set_ylabel("Petal length (cm)")
chart3.set_title("Petal length vs class")

chart4.scatter(dataset['flower_class'],dataset['petal_wid'])
chart4.set_xlabel("Class of flower")
chart4.set_ylabel("Petal width (cm)")
chart4.set_title("Petal width vs class")

plt.show()
```

```python
figure2 = plt.figure(1, figsize=(20, 10))

chart1 = figure2.add_subplot(221)
chart2 = figure2.add_subplot(222)
chart3 = figure2.add_subplot(223)
chart4 = figure2.add_subplot(224)

chart1.hist(dataset['sepal_len'], bins= 10)
chart1.set_xlabel("Sepal length (cm)")
chart1.set_title("Sepal length freq")

chart2.hist(dataset['sepal_wid'], bins= 10)
chart2.set_xlabel("Sepal width (cm)")
chart2.set_title("Sepal width freq")

chart3.hist(dataset['petal_len'], bins= 10)
chart3.set_xlabel("Petal length (cm)")
chart3.set_title("Petal length freq")

chart4.hist(dataset['petal_wid'], bins= 10)
chart4.set_xlabel("Petal width (cm)")
chart4.set_title("Petal width freq")

```
