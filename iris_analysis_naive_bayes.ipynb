```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print("Naive Bayes Classification of Iris Data")

#reading datasets
iris_data = pd.read_csv("iris_dataset.csv")

#segregating dataset into input(attributes) and output(target variable) datasets
#here target variable is flower_class
X = iris_data.drop(columns=['flower_class'], axis=1)
y = iris_data['flower_class']

#spliting input and output datasets for testing n training
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)

#creating model 
model = GaussianNB()

#training it 
model.fit(x_train, y_train)

#predict target variable (flower_class) based on the attributes in input dataset
#predict target on training dataset
train_predictions = model.predict(x_train)
print("\nPredictions made for flower_class on training dataset: \n", train_predictions)

#accuracy score on training dataset
accuracy_train = accuracy_score(y_train,train_predictions)
print("\nAccuracy on train dataset: ", accuracy_train)

#prediction on testing set
test_predictions = model.predict(x_test)
print("\nPredictions made for flower_class on testing dataset: \n", test_predictions)

#accuracy score on testing dataset
accuracy_test = accuracy_score(y_test,test_predictions)
print("\nAccuracy on test dataset: ", accuracy_test)
```
